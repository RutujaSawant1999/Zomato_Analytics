# Step 1 -Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 - Create the data frame

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# Convert the datatype of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# What type of restaurant do the majority of customers order from ?

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restraunt") 
```

## Conclusion - Majority of the restraunt falls in dining category 

# How many votes has each type of restraunt received from customers ?

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel("Type of restraunt", c="red", size=20)
plt.ylabel("Votes", c="red",size=20)
```

## Conclusion - Dining restraunts has received maximum votes

# What are the ratings that the majority of restraunts have received ?

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins = 5)
plt.title("ratings distribution")
plt.show()
```

## Conclusion - The majority restraunts received ratings from 3.5 to 4

# Zomato has observed that most couples order most of their food online. What is their average spending on each order ? 

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

## Conclusion - The majority of couples prefer restraunts with an approximate cost of 300 rupees

# Which mode(online or offline) has received the maximum rating ?

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order' , y ='rate', data= dataframe)
```

## Conclusion - Offline order received lower rating in comparison to online order

# Which type of restraunt received more offline orders, so that Zomato can prefer customers with some good offers ?

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

## Conclusion - Dining restraunts primarily accept offline orders, whereas cafes primarily receivesonline orders.This suggests that clients prefer to orders in person at restraunts, but prefer online ordering at cafes.
